```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler, KBinsDiscretizer
from sklearn.compose import ColumnTransformer
```

```python
msft = yf.Ticker("MSFT")
history = msft.history(period="720d")
closes = history["Close"]
media_movel = closes.rolling(window=10).mean()
retornos = closes.pct_change().round(4)
desvio_padrao = closes.std()
print(desvio_padrao)
```
